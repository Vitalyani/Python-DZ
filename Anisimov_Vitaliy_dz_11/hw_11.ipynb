```python
# Дана функция:
# f(x) = x^2 - 6*abs(x) + 8
```

```python
from sympy import *
from sympy.plotting import plot
init_printing()
x = Symbol('x', real=True)
y = x**2 - 6*abs(x) + 8
y
```

Найдем корни уравнения

```python
# 1. Определить корни
result = solve(y)
result
```

Найдем производную 

```python
dif_y = diff(y)
dif_y
```

Строим график производной

```python
plot(dif_y)
```

```python
result_dif_y = solve(dif_y)
result_dif_y
```

Добавим вручную остальные корни х, которые видно на графике

```python
result_dif_y = solve(dif_y)
all_roots = [-3, 0, *result_dif_y]
all_roots
```

```python
# 2. Найти интервалы, на которых функция возрастает
solve(dif_y > 0)
```

```python
# 3. Найти интервалы, на которых функция убывает
solve(dif_y < 0)
```

```python
# 4. Построить график
plot(y)
```

```python
# 5. Вычислить вершину(-ы)
graph_vertices = []
for peak in result_dif_y:
    graph_vertices.append((peak, y.subs(x, peak)))
graph_vertices
```

Если взять все кони х, которые добавляли в ручную, то...

```python
graph_vertices = []
for peak in all_roots:
    graph_vertices.append((peak, y.subs(x, peak)))
graph_vertices
```

Вершины можно увидеть по первому грфику

```python
# 6. Определить промежутки, на котором f > 0
solve(y > 0)
```

```python
# 7. Определить промежутки, на котором f < 0
solve(y < 0)
```
